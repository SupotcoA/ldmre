```python
import matplotlib.pyplot as plt
import numpy as np

rho = 7
sigma_min = 0.002
sigma_max = 80
n_steps=512
t = np.arange(0,n_steps+1)
t = (sigma_max**(1/rho)+\
        t/n_steps*\
        (sigma_min**(1/rho)-\
        sigma_max**(1/rho)))**rho

plt.plot(t)
plt.plot(-np.diff(t))
plt.show()
```

```python
plt.plot(-np.diff(t)/t[:-1])
```
